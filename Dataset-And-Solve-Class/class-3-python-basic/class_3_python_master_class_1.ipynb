## 3. Variables and Data Types
Variables are containers for storing data values. Think of them like "Gift Boxes" where you store your data.

### Rules for Naming Variables:
1.  Must start with a letter (a-z, A-Z) or underscore (_).
2.  Cannot start with a number.
3.  Can only contain alpha-numeric characters and underscores.
4.  Case-sensitive (`age`, `Age` and `AGE` are different variables).
5.  Cannot use Python reserved keywords (e.g., `if`, `else`, `print`).

```python
# Variable Assignment
my_message = "My message to the world"
print(my_message)

# Reassigning variables (Python is dynamically typed)
x = 10
print("Initial x:", x)
x = 20
print("New x:", x)
```

```python
# Multiple Assignment
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")

# Swapping variables (The Pythonic Way)
x = 5
y = 10
print(f"Before Swap: x={x}, y={y}")

x, y = y, x
print(f"After Swap: x={x}, y={y}")
```

```python
# Deleting a variable to free up memory
temp = "Temporary Data"
del temp
# print(temp) # This will raise a NameError because temp no longer exists
```

### Data Types
* **Integer (`int`):** Whole numbers (e.g., 50).
* **Float (`float`):** Decimal numbers (e.g., 3.14).
* **String (`str`):** Text data (e.g., "Hello").
* **Boolean (`bool`):** Logic values (True/False).

```python
# Checking Data Types
my_int = 50
print("50 is:", type(my_int))

my_float = 3.14
print("3.14 is:", type(my_float))

my_str = "BongoDev"
print("'BongoDev' is:", type(my_str))

my_bool = True
print("True is:", type(my_bool))
```

## 4. Type Casting
Type casting is the process of converting one data type into another.

```python
# Converting types
my_bool_val = True
my_float_val = 3.99
my_int_string = '412578'

# Boolean to Int (True is 1, False is 0)
print(f"Bool to Int: {int(my_bool_val)}")

# Float to Int (Truncates decimal part)
print(f"Float to Int: {int(my_float_val)}")

# String to Float
print(f"String to Float: {float(my_int_string)}")
```

## 5. Operators

### Arithmetic Operators
* `+` (Addition)
* `-` (Subtraction)
* `*` (Multiplication)
* `/` (True Division - returns float)
* `//` (Floor Division - returns integer part)
* `%` (Modulo - returns remainder)
* `**` (Exponentiation - power)

```python
print(f"10 + 5 = {10 + 5}")
print(f"10 / 2 = {10 / 2}")   # Returns 5.0
print(f"10 // 3 = {10 // 3}") # Returns 3 (quotient)
print(f"10 % 3 = {10 % 3}")   # Returns 1 (remainder)
print(f"2 ** 3 = {2 ** 3}")   # 2 to the power of 3
```

### Relational & Logical Operators
Used for comparison and boolean logic.

```python
# Relational (Returns True/False)
print(f"Is 4 equal to 4? {4 == 4}")
print(f"Is 4 not equal to 5? {4 != 5}")

# Logical (and, or, not)
print(f"True and False: {2 < 4 and 2 > 6}")
print(f"True or False: {2 < 4 or 2 > 6}")
print(f"Not False: {not False}")
```

```python
# Bitwise Shift Operators (Advanced)
# Left Shift (<<): Multiplies by 2^n
a = 2
b = 3
print(f"2 << 3 (2 * 2^3): {a << b}")

# Right Shift (>>): Divides by 2^n (integer division)
a = 25
b = 3
print(f"25 >> 3 (25 // 2^3): {a >> b}")
```

## 6. Conditional Statements
Control the flow of the program based on conditions.

```python
age = 25
income = 50000

# Checking eligibility
if age > 18 and income > 30000:
    print("You are eligible for the program.")
else:
    print("You are not eligible.")
```

```python
# Elif Ladder (Grading System)
grade = 85

if grade >= 90:
    print("Grade: A")
elif grade >= 80:
    print("Grade: B")
elif grade >= 70:
    print("Grade: B+")
else:
    print("Grade: C")
```
