```python
"""
Understanding the categories of visual plots

Statistical plots:
1. Relational plot (relplot): establishes relations between variables
   x = 1, y = 2
   x = 2, y = 4
   x = 3, y = 6
   independent variable: the values of the variable does not depend on other variables i.e. x
   dependent variable: the value of the variable depends on other variables i.e. y is a dependent variable as it changes based on x. 
2. Distributional plot (displot): visualizes the distribution of variables
3. Categorical plot (catplot): provides statistical information about the categorical variables

matplotlib > seaborn > plotly
"""

sns.get_dataset_names()
```

```python
""" Data Visualization using Seaborn """
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')
penguins = sns.load_dataset('penguins')
```

```python
tips.head()
```

# Relational plots

```python
""" Scatter Plot 

Statistics:

Correlation between variables
   [-1, 1]
   -1: Strongly negatively correlated 
   0: Not correlated
   1: Strongly positively correlated
"""

plot = sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
)

__ = plot.set(
    xlabel='Total Bill (USD)',
    ylabel='Tip (USD)',
    title='Relation between Total Bill and Tip (USD)'
)
```

```python
tips['day'].value_counts()
```

```python
plot = sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='day',
    row='day',
    height=4,
)
```

```python
""" Line plot 
- to visualize the changes of a variable with respect to another variable.
"""

plot = sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    height=4,
    col='region',
    hue='region',
)

plot.set(
    xlabel='Time plot', 
    ylabel='Signal'
)
```

# Distributional Plot

```python
"""
Event: Waiting for a bus

 Day 1, wait time = 1 min
 Day 2, wait time = 2 min
 Day 3, wait time = 6 min
 Day 4, wait time = 3 min
 Day 5, wait time = 15 min
 Day 6, wait time = 8 min
 
Average waiting time: (1+2+6+3+15+8)/6 = 5.833 min

"""
penguins.head()
```

```python
""" Distributional plot: Flipper length 

- Univariate distribution: Distribution of one variable/feature/column
- Bivariate distribution: Distribution of two variables/features/columns
- Multivariate distribution: Distribution of three or more variables/features/columns
"""

sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='hist',
    bins=20,
)
```

```python
sns.displot(
    data=penguins,
    x='species',
    discrete=True,
    shrink=0.8,
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step',
    multiple='stack',
    stat='probability',
)
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    kind='kde', # kernel density estimation
    fill=True,
    hue='species',
    multiple='stack'
)
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    kind='kde', # kernel density estimation
    fill=True,
)
```

```python
sns.displot(
    data=penguins, 
    x="bill_length_mm", 
    y="bill_depth_mm",
    cbar=True
)
```

```python
sns.pairplot(penguins)
```

# Categorial Plot

```python
""" Strip plot 
- It shows individual data points as a scatter plot along categorical axis.
- It visualizes outliers, clusters, and distribution.
- Limitations: data can overlap, making it hard to understand the distribution.
"""

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='strip',
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    jitter=False,
    order=['Sun', 'Sat', 'Thur', 'Fri']
)
```

```python
""" Swarm plot 
- It is similar to strip plot but adjust the points to avoid overlap.
- It can visualize outliers, clusters, and distribution for small dataset
- Limitation: Does not fit well for larger datasets for spacing issues.
"""

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='swarm',
)
```

```python
""" Box Plot 

Statistics:
 minimum, Q1 (25%), median, Q3(75%), maximum
 
- It shows distribution through statistics
- It visualizes outliers, distributions.
- Limitations: No information present for understanding data density.
"""

sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="box",
)
```

```python
""" Violine plot 

- It combines box plot and distribution density graph (KDE: Kernel Density Estimation)
- It visualizes data distribution and statistics
"""

sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="violin",
)
```

```python
""" Data visualization with plotly """

""" scatter plot with plotly 

sns            |   px
===============================
kind='scatter' | px.scatter
data           | data_frame
hue            | color
col            | facet_col
row            | facet_row

"""
import plotly.express as px

plot = px.scatter(
    data_frame=tips,
    x='total_bill',
    y='tip',
    color='sex',
    facet_col='time',
    symbol='smoker'
)

plot.update_layout(
    xaxis_title='Total Bill (USD)',
    yaxis_title='Tip (USD)',
    title='Total Bill Vs Tip',
)

plot.show()
```
