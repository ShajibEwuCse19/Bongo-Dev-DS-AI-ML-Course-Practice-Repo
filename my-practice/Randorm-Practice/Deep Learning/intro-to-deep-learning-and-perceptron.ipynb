````python
# Neural Network Parameter
nn.Parameter(torch.randn(3)) # requires_grad=True -> during training, these parameters will be optimized and calculate gradient
````

````python
#Define Perceptron, we need input(x), weights(w), bias(b), calculation(z), activation function
class Perceptron(nn.Module):
    def __init__(self, input_size):
        super(Perceptron, self).__init__() #We want the features of nn.Module in our parent class Perceptron. Without this, PyTorch will not understand
                                           # that it's a neural network

        self.input_size = input_size #store total no of input(x) size or no of features in each input

        self.weight = nn.Parameter(torch.randn(input_size)) #we know, all input (X) has an corresponding weigh(W). so, input_size == no of weight
        self.bias = nn.Parameter(torch.zeros(1)) #Without bias (b), all result will went through origin. bias create flexibity [can shift anywhere]
                                                 #nn.Parameter() -> Trainable Parameters (will apply optimizer, gradient, etc.)

        print(self.weight, self.bias)

    def forward(self, x): #Here, we'll write the prediction logics
        z = x @ self.weight + self.bias #@ means matrix multiplication / dot product and z = X.W + b = x1w1  + x2w2 + ---- + xnwn + b
        y_pred = torch.sigmoid(z) #push/insert the result(z) to a non-linear activation function 'sigmoid'. Introduce non-linearity.
                                  #Why sigmoid? -> Useful for binary classification. 0 → negative and 1 → positive
        return y_pred

# When ever well call perceptron(x), internally PyTorch calls forward(x)

perceptron = Perceptron(3) #here, x = [x1, x2, x3] -> 3 features
````

````python
x = torch.tensor([1.0, 2.0, 3.0])
output = perceptron(x) #When ever we'll call perceptron(x), internally PyTorch calls forward(x)
print(output)
````

````python
#Define model
import os

model = Perceptron(input_size=3) #input_size = 3 means 3 features in each input
model_path = 'my_model.pth' #define a path for save our model

if os.path.exists(model_path):
    model.load_state_dict(torch.load(model_path)) #load if there is any existing learned model. Sometimes we face interruped and we need to restart. 
                                                  #So, using this method, we can start with existing situation so that we don't need to start from beginning.
    print("Model Loaded...")
````

````python
#input(x)
x = torch.randn(3)
print(x)
````

````python
#prediction/output(y)
y_pred = model(x)
print(y_pred)
````

````python
#see the parameters of our model
print(*model.parameters())
````

````python
# Cost Function
criterion = nn.BCELoss() #Bynary Cross Entropy Loss - used for binary classification

#Optimizer - used for optimize the loss/cost
optimizer = torch.optim.Adam(
    model.parameters(), #we need to optimize the model's parameters
    lr=0.0001 #define learning rate
)
````

````python
# select input (X), output (Y)
X = torch.Tensor(
    [
        [0.0, 0.0, 0.0], # sample input 1 (x1, x2, x3)
        [0.0, 1.0, 0.0], # sample input 2
        [1.0, 0.0, 0.0], # sample input 3
        [1.0, 1.0, 1.0]  # sample input 4
    ]
) #shape (4, 3) -> Row/Batch size 4 (total input), Col 3 (no of features in each sample). Batch processing: (4,3)⋅(3,)=(4,)
# (batch_size, features) = (4,3)

Y = torch.Tensor(
    [
        [0.0], # sample output 1
        [0.0], # sample output 2
        [0.0], # sample output 3
        [1.0]  # sample output 4
    ]
)
````

````python
y_pred = model(X) # means perceptron.forward(X)
print(y_pred)
````

````python
print(y_pred.ndim)
print(y_pred)
y_pred = y_pred.unsqueeze(axis=1) #unsqueeze() -> add a dimension. We did this for creating a batch output. 
                                  #eural networks standardize tensors as (batch_size, output_features)
print(y_pred.ndim)
print(y_pred)
y_pred = y_pred.unsqueeze(axis=1) # unsqueeze() -> add a dimension
print(y_pred.ndim)
print(y_pred)
````

````python
#model training
def train_model(model, epochs):
    for epoch in range(epochs):
        """ Prediction (Forward Propagation) """
        y_pred = model(X) # (Forward propagation) -> produce input to output
        
        """ Loss calculation """
        loss = criterion(y_pred.unsqueeze(axis=1), Y) # calculate loss using predict output(y_pred) and true output Y (original)

        """ Calculate gradient """
        optimizer.zero_grad() #finding the optimal points [local minima] where loss is minimum
        loss.backward() # (Backward propagation)

        """ Update weight """
        optimizer.step() # w1 = w1 - lr * dw1, here, lr = steps, dw = direction
        
        if epoch % 1000 == 0:
            print(f'Epoch: {epoch}, Loss: {loss.item():.4f}')
````

````python
train_model(model, 5000)
````

````python
print(*model.parameters())
````

````python
#Model Evaluation
""" Mode of a torch model

By default: train
We can set the mode to eval during prediction
"""
def predict(model, X):
    model.eval() # without evaluation mode, it will calculate gradient which is very time consuming and unnecessary. 
    with torch.no_grad():
        test_output = model(X)
    return test_output

predict(model, X)
````

## 2. Perceptron কী?

Perceptron হলো neural network-এর সবচেয়ে basic unit। এটি input নেয়, weight দিয়ে multiply করে, bias add করে, তারপর activation function দিয়ে output দেয়।

### Formula

যদি input হয়:

\[x = [x_1, x_2, x_3]\]

weight হয়:

\[w = [w_1, w_2, w_3]\]

bias হয়:

\[b\]

তাহলে perceptron-এর linear output:

\[z = x_1w_1 + x_2w_2 + x_3w_3 + b\]

vector form-এ:

\[z = XW + b\]

তারপর sigmoid activation:

\[\hat{y} = \sigma(z) = \frac{1}{1 + e^{-z}}\]

এখানে `ŷ` হচ্ছে predicted probability।

## 4. Perceptron Class Explanation

### Cell 9
এটি notebook-এর core অংশ।

```python
class Perceptron(nn.Module):
```
এখানে custom neural network model define করা হয়েছে। `nn.Module` inherit করার কারণ হলো PyTorch যেন এটিকে proper model হিসেবে বুঝতে পারে।

### `__init__` method
`input_size` বোঝায় প্রতি sample-এ কতটি feature আছে।

```python
self.weight = nn.Parameter(torch.randn(input_size))
```
এখানে input-এর প্রতিটি feature-এর জন্য একটি weight রাখা হয়েছে। যদি `input_size = 3` হয়, তাহলে weight হবে `[w1, w2, w3]`।

```python
self.bias = nn.Parameter(torch.zeros(1))
```
Bias model-কে flexibility দেয়। Bias না থাকলে model-এর decision boundary origin-এর সাথে অতিরিক্তভাবে depend করতে পারে।

### `forward()` method
```python
z = x @ self.weight + self.bias
```
এখানে dot product করা হচ্ছে।

যদি:
\[x = [1,2,3], \; w = [0.5,-1,2], \; b = 0.1\]
তাহলে:
\[z = 1(0.5) + 2(-1) + 3(2) + 0.1 = 4.6\]

তারপর:
```python
y_pred = torch.sigmoid(z)
```
Sigmoid output-কে 0 থেকে 1 এর মধ্যে নিয়ে আসে। Binary classification-এ এটি probability হিসেবে useful।

Sigmoid formula:
\[\sigma(z) = \frac{1}{1 + e^{-z}}\]

যদি `z = 4.6` হয়, তাহলে output প্রায় `0.99` হবে। মানে model class 1-এর প্রতি খুব confident।

### `perceptron = Perceptron(3)`
এখানে 3-feature input-এর জন্য perceptron instance তৈরি করা হয়েছে।

### Cell 15
এটি training-এর core setup:

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

#### `BCELoss` কী?
Binary Cross Entropy loss binary classification-এর জন্য ব্যবহৃত হয়।

Formula:
\[L = -[y\log(\hat{y}) + (1-y)\log(1-\hat{y})]\]

এখানে:
- `y` = actual label
- `ŷ` = predicted probability

যদি actual = 1 এবং prediction = 0.99 হয়, loss কম হবে। কিন্তু actual = 1 আর prediction = 0.01 হলে loss অনেক বেশি হবে।

#### `Adam` optimizer কী?
Adam একটি widely used optimizer, যা gradient ব্যবহার করে weight update করে। এটি adaptive learning rate ব্যবহার করে এবং practical deep learning-এ খুব common।

Gradient descent intuition:
\[w_{new} = w_{old} - \eta \frac{\partial L}{\partial w}\]
এখানে `η` হলো learning rate।

## 7. Training Loop, Backpropagation, Evaluation, Save/Load

### Cell 19
এই cell-এ `train_model()` function define করা হয়েছে। এটি পুরো deep learning training cycle-এর সবচেয়ে গুরুত্বপূর্ণ অংশ।

### Step 1: Forward Propagation
```python
y_pred = model(X)
```
Input `X` model-এর মধ্যে যায় এবং prediction আসে।

### Step 2: Loss Calculation
```python
loss = criterion(y_pred.unsqueeze(axis=1), Y)
```
Prediction আর actual label compare করে error বের করা হচ্ছে।

### Step 3: Old Gradient Clear
```python
optimizer.zero_grad()
```
PyTorch default-এ gradient accumulate করে। তাই প্রতিটি নতুন epoch-এর আগে old gradient clear করা জরুরি।

### Step 4: Backpropagation
```python
loss.backward()
```
এখানে autograd loss থেকে weight ও bias-এর gradient বের করে। অর্থাৎ:
\[\frac{\partial L}{\partial w}, \frac{\partial L}{\partial b}\]

এগুলো বলে দেয় loss কমাতে parameter কোন দিকে update করতে হবে।

### Step 5: Weight Update
```python
optimizer.step()
```
Optimizer gradient ব্যবহার করে parameter update করে।

Update intuition:
\[w = w - \eta \frac{\partial L}{\partial w}\]
\[b = b - \eta \frac{\partial L}{\partial b}\]

প্রতি 1000 epoch পর loss print করা হচ্ছে, যাতে training progress monitor করা যায়। Loss কমলে বুঝতে পারি model pattern শিখছে।

### Tensor
Deep learning-এর main data structure।

### Parameter
Trainable values, যেমন weight ও bias।

### Forward Pass
Input থেকে output prediction আসা।

### Activation Function
Sigmoid non-linearity introduce করে এবং output-কে probability range-এ আনে।

### Loss Function
Prediction কতটা ভুল হয়েছে তা মাপে।

### Backpropagation
Loss থেকে gradient বের করে।

### Optimizer
Gradient ব্যবহার করে parameter update করে।

### Evaluation Mode
Training ছাড়া efficient prediction করার mode।

### Model Save/Load
Learned parameter future use-এর জন্য file-এ সংরক্ষণ করা।
