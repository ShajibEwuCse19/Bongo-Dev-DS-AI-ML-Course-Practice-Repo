```python
import seaborn as sns 
sns.__version__
```

```python
sns.set(
    style='whitegrid',
    palette='flare',
    font='Arial',
    font_scale=1
)
```

```python
# buildin Daset list
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins')
fmri = sns.load_dataset('fmri')
```

```python
tips.head()
```

```python
plot = sns.relplot(
    data = tips,
    x='total_bill',
    y='tip',
    kind='scatter'
)
plot.set(
    title='Relation Between Total Bill and Tip',
    xlabel='Total Bill (USD)',
    ylabel='Tip (USD)',
)
```

```python
tips['day'].value_counts()
```

```python
plot = sns.relplot(
    data = tips, #dataset load
    x='total_bill', #x axis
    y='tip', #y axis
    kind='scatter', #what kind of plot. scatter, line, etc.
    hue='day', #color based on category, here, day wise color
)
plot.set(
    title='Relation Between Total Bill and Tip',
    xlabel='Total Bill (USD)',
    ylabel='Tip (USD)',
)
```

Scatter Plot - Relational Plot

```python
from pyparsing import col
plot = sns.relplot(
    data = tips, #dataset load
    x='total_bill', #x axis
    y='tip', #y axis
    kind='scatter', #what kind of plot. scatter, line, etc.
    hue='day', #color based on category, here, day wise color
    col='day', #create color for every day inside dataset
    #row='day', #same as col
    height=4, #height maintain [linke zoom in out]
)
plot.set(
    title='Relation Between Total Bill and Tip',
    xlabel='Total Bill (USD)',
    ylabel='Tip (USD)',
)
```

```python
plot = sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='event'
).set(
    title='Relation between time point and signal for FMRI data',
    xlabel='Time Point',
    ylabel='Signal',
)
```

```python
plot = sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='region',
    col='region'
).set(
    title='Relation between time point and signal for FMRI data',
    xlabel='Time Point',
    ylabel='Signal',
)
```

Distribultion Plot

```python
penguins.head()
```

```python
plot = sns.displot(
    data=penguins,
    x='flipper_length_mm',
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
plot = sns.displot(
    data=penguins,
    x='flipper_length_mm',
    bins=8,
    kind='hist', #default histogram
    shrink=0.8, #decrease size, max = 1.0
    hue='species'
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
plot = sns.displot(
    data=penguins,
    x='flipper_length_mm',
    bins=8,
    shrink=0.8, #decrease size, max = 1.0
    hue='species',
    col='species',
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    bins=8,
    shrink=0.8,
    hue='species',
    element='step', #step by step
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    bins=8,
    shrink=0.8,
    hue='species',
    element='step', #step by step
    multiple='stack'
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    bins=8,
    shrink=0.8,
    hue='species',
    kind='hist', #default histogram
    element='step', #step by step
    multiple='stack',
    stat='probability', #remove count, add probability [percentage if total is 100%]
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Probability',
)
```

KDE - Kerner Density Estimation

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='kde', 
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='kde', 
    hue='species',
    multiple='stack',
).set(
    title='Histogram of flipper_length_mm',
    xlabel='flipper_length (mm)',
    ylabel='Count',
)
```
