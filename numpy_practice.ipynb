```python
import numpy as np
np.__version__
```

```python
my_list = [1, 2, 3, 4, 5]
my_np_array = np.array(my_list)
print(my_list, type(my_list))
print(my_np_array, type(my_np_array))
```

```python
a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

```python
arr = np.array([1, 2, 3], ndmin=5)
print(arr)
print('number of dimensions :', arr.ndim)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print("2nd element of first row: ", arr[0,1])
print("2nd element of first row: ", arr[0][1])
```

```python
arr = np.array([
    [[1, 2, 3], [4, 5, 6]], #first array
    [[7, 8, 9], [10, 11, 12]] #second array
    ])
print(arr.shape)
print("2nd element of 4th row = ", arr[1, 1, 1])
```
